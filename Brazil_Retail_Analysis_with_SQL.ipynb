```python
import pandas as pd

# Load CSV files into DataFrames
file_name1 = 'olist_customers_dataset.csv'
file_name2 = 'olist_geolocation_dataset.csv'
file_name3 = 'olist_order_items_dataset.csv'
file_name4 = 'olist_order_payments_dataset.csv'
file_name5 = 'olist_order_reviews_dataset.csv'
file_name6 = 'olist_orders_dataset.csv'
file_name7 = 'olist_products_dataset.csv'
file_name8 = 'olist_sellers_dataset.csv'
file_name9 = 'product_category_name_translation.csv'

df_customers = pd.read_csv(file_name1)
df_geolocation = pd.read_csv(file_name2)
df_order_items = pd.read_csv(file_name3)
df_order_payments = pd.read_csv(file_name4)
df_order_reviews = pd.read_csv(file_name5)
df_orders = pd.read_csv(file_name6)
df_products = pd.read_csv(file_name7)
df_sellers = pd.read_csv(file_name8)
df_product_categories = pd.read_csv(file_name9)
```

```python
import sqlite3

# Create a new SQLite database (or connect to one if it exists)
conn = sqlite3.connect('olist_ecommerce.db')

# Write DataFrames to SQL tables
df_customers.to_sql('customers', conn, if_exists='replace', index=False)
df_geolocation.to_sql('geolocation', conn, if_exists='replace', index=False)
df_order_items.to_sql('order_items', conn, if_exists='replace', index=False)
df_order_payments.to_sql('order_payments', conn, if_exists='replace', index=False)
df_order_reviews.to_sql('order_reviews', conn, if_exists='replace', index=False)
df_orders.to_sql('orders', conn, if_exists='replace', index=False)
df_products.to_sql('products', conn, if_exists='replace', index=False)
df_sellers.to_sql('sellers', conn, if_exists='replace', index=False)
df_product_categories.to_sql('product_categories', conn, if_exists='replace', index=False)
```

**1.Customers to Orders: Join olist_customers_dataset with olist_orders_dataset on customer_id.**




```python
#Customers to Orders: Join olist_customers_dataset with olist_orders_dataset on customer_id
q = """
SELECT customers.customer_id, customers.customer_unique_id, orders.order_id
FROM customers
JOIN orders ON customers.customer_id = orders.customer_id;
"""
pd.read_sql_query(q, conn)
```

#1. Customer Orders

```python
#Customer Orders
#Now let us get the total number of the orders per customer
q = """
SELECT customers.customer_unique_id, COUNT(orders.order_id) AS total_orders
FROM customers
JOIN orders ON customers.customer_id = orders.customer_id
GROUP BY customers.customer_unique_id
ORDER BY total_orders DESC;
"""
pd.read_sql_query(q, conn)
```

#2. Top Products Sold

```python
#Now let us get the top products being sold
q = """
SELECT products.product_category_name, COUNT(order_items.order_item_id) AS total_sold
FROM products
JOIN order_items ON products.product_id = order_items.product_id
GROUP BY products.product_category_name
ORDER BY total_sold DESC;
"""
pd.read_sql_query(q, conn)
```

#3. Average Review Scores by Product

```python
#Average Review Scores by Product
q = """
SELECT products.product_category_name, AVG(order_reviews.review_score) AS avg_review_score
FROM products
JOIN order_items ON products.product_id = order_items.product_id
JOIN orders ON order_items.order_id = orders.order_id
JOIN order_reviews ON orders.order_id = order_reviews.order_id
GROUP BY products.product_category_name
ORDER BY avg_review_score DESC;
"""
pd.read_sql_query(q, conn)
```

```python
#Products with average review scores more than 4
q = """
SELECT products.product_category_name, AVG(order_reviews.review_score) AS avg_review_score
FROM products
JOIN order_items ON products.product_id = order_items.product_id
JOIN orders ON order_items.order_id = orders.order_id
JOIN order_reviews ON orders.order_id = order_reviews.order_id
GROUP BY products.product_category_name
HAVING avg_review_score > 4
ORDER BY avg_review_score DESC;
"""
pd.read_sql_query(q, conn)
```

# Advanced Analysis

**1. Sales Trends**

Investigating the number of orders over time, revenue, or sales by product category.

```python
#Investigating the number of orders over time, revenue, or sales by product category.
q = """
SELECT orders.order_purchase_timestamp, SUM(order_items.price) AS total_revenue
FROM orders
JOIN order_items ON orders.order_id = order_items.order_id
GROUP BY orders.order_purchase_timestamp
ORDER BY orders.order_purchase_timestamp;
"""
pd.read_sql_query(q, conn)
```

```python
#Can I get the dates with the highest revenue
q = """
SELECT orders.order_purchase_timestamp, SUM(order_items.price) AS total_revenue
FROM orders
JOIN order_items ON orders.order_id = order_items.order_id
GROUP BY orders.order_purchase_timestamp
ORDER BY total_revenue DESC;
"""
pd.read_sql_query(q, conn)
```

```python
#The most grossing product categories
q = """
SELECT products.product_category_name, SUM(order_items.price) AS total_revenue
FROM products
JOIN order_items ON products.product_id = order_items.product_id
GROUP BY products.product_category_name
ORDER BY total_revenue DESC;
"""
pd.read_sql_query(q, conn)
```

```python
#- Total revenue generated over time (monthly)
q = """
SELECT strftime('%Y', orders.order_purchase_timestamp) AS year,
       SUM(order_items.price) AS total_revenue
FROM orders
JOIN order_items ON orders.order_id = order_items.order_id
GROUP BY year
ORDER BY year;
"""
df = pd.read_sql_query(q, conn)
df.head()
```

#5. Average Order Value by Month

Measure the average value of an order over time.

```python
#Meaure the average value o an order over time
q = """
SELECT strftime('%Y-%m', orders.order_purchase_timestamp) AS year_month,
       AVG(order_items.price) AS avg_order_value
FROM orders
JOIN order_items ON orders.order_id = order_items.order_id
GROUP BY year_month
ORDER BY year_month;
"""
pd.read_sql_query(q, conn)
```

#6. Customer Lifetime Value (CLTV)

Estimate how much each customer is worth based on their total spending.




```python
#Estimate how much each customer is worth based on their total spending.
#-- Customer lifetime value (CLTV)
q = """
SELECT
    c.customer_id,
    SUM(oi.price) AS lifetime_value
FROM
    customers c
JOIN
    orders o ON c.customer_id = o.customer_id
JOIN
    order_items oi ON o.order_id = oi.order_id
GROUP BY
    c.customer_id
ORDER BY
    lifetime_value DESC;
""" # Removed the extra double quote at the beginning
pd.read_sql_query(q, conn)
```
